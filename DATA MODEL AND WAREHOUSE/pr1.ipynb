```python
def create_database():
    conn=psycopg2.connect("host=127.0.0.1 dbname=postgres user=postgres password=postgres")
    cur=conn.cursor()
    conn.set_session(autocommit=True)
    #create database
    cur.execute("DROP DATABASE IF EXISTS energy_data")
    cur.execute('CREATE DATABASE energy_data')
    conn.close()
    conn=psycopg2.connect("host=127.0.0.1 dbname=energy_data user=postgres password=postgres")
    cur=conn.cursor()
    return conn,cur
    
```

```python
conn,cur=create_database()
```

```python
table_t=("""create table if not exists t_table(
T_out float PRIMARY KEY,
T_1 FLOAT,
T_2 FLOAT,
T_3 FLOAT,
T_4 FLOAT,
T_5 FLOAT,
T_6 FLOAT,
T_7 FLOAT,
T_8 FLOAT,
T_9 FLoaT
)""")
cur.execute(table_t)
conn.commit()
```
